# Positional Encondings

See: https://kazemnejad.com/blog/transformer_architecture_positional_encoding/
See: https://timodenk.com/blog/linear-relationships-in-the-transformers-positional-encoding/

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
d = 128 # dimensionality of the feature vector/position encoding
N = 10000 # controls the steepness
```

```python
def embedding_vector(position):

    emb = []    
    for i in range(d):
        
        k = i//2
        omega = 1/math.pow(N,2*k/d)
        if i%2==0:
            emb.append(math.sin(omega*position))
        else:
            emb.append(math.cos(omega*position))
    
    return np.array(emb)
```

```python
for p in [0,1,5,10,50]:
    emb = embedding_vector(p)
    plt.plot(emb, label=p)
    plt.ylim([-1,1])
    plt.legend()
    plt.show()
```

```python
max_position = 100

embs = []
for p in np.arange(max_position):
    embs.append(embedding_vector(p))
embs = np.stack(embs)

plt.figure(figsize=(15,15))
plt.imshow(embs, cmap='PiYG')
plt.colorbar(shrink=0.3)
plt.title(f'The first {max_position} positions')
plt.show()
```
