```python
from micrograd import *
```

```python
# Forward pass

x = Value(2.0, label="x"); y = Value(3.0, label="y")
z = x + y; z.label = "z"
a = z.tanh(); a.label = "a"
b = a * x; b.label = "b"

draw_dot(b)
```

```python
# Backward pass

b.grad = 1.0; b._backward()
a._backward()
z._backward()
x._backward(); y._backward() # Will not do anything since their backward functions are empty

draw_dot(b)
```
